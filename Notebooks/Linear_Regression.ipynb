## <font color="light blue">Simple Linear Regression </font>

#### <font color="orange">What is Simple Linear Regression? </font>

Simple Linear Regression is a statistical method that helps us understand the relationship between two continuous variables. It's called "simple" because we're only dealing with two variables, and "linear" because we're looking for a straight-line relationship between them.

##### **<font color="orange">Example </font>**

Let's say we want to predict the price of a house based on its size. We have data on the size of several houses (in square feet) and their corresponding prices. We can use Simple Linear Regression to find the relationship between these two variables.

#### <font color="orange">Mathematical Equation </font>

The Simple Linear Regression equation is:

Price(y) = β0 + β1 \* Size(x) + ε

Where:

* **Price** is the predicted price of the house
* **Size** is the size of the house (in square feet)
* **β0** is the intercept or constant term
* **β1** is the slope coefficient - unit movement in the x-axis what is the unit movement in the y-axis.
* **ε** is the error term (which represents the random variation in the data)

#### How it Works

Here's a step-by-step explanation:

1. **Collect data**: We collect data on the size and price of several houses.
2. **Plot the data**: We plot the data on a scatter plot to visualize the relationship between size and price.
3. **Find the best-fit line**: We use a mathematical algorithm to find the best-fit line that minimizes the error between the predicted prices and the actual prices.
4. **Interpret the results**: We interpret the results by looking at the slope coefficient (β1) and the intercept (β0). The slope coefficient tells us the change in price for a one-unit change in size. The intercept tells us the predicted price when the size is zero.

#### <font color="orange">Cost Function </font>

A cost function, also known as a loss function or objective function, is a mathematical function that measures the difference between the predicted output of a machine learning model and the actual true output.

The cost function is typically denoted as J(θ), where θ represents the model's parameters. The cost function takes the following form:
The formula for the cost function is:

$$
J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})^2
$$

* $J(\theta)$: Cost function
* $\theta$: Model parameters
* $m$: Number of training examples
* $x^{(i)}$: $i^{th}$ input feature
* $y^{(i)}$: $i^{th}$ actual true output
* $h_\theta(x^{(i)})$: Predicted output of the model for the $i^{th}$ datapoint


<font color = 'red'>**Our main aim is to minimize the cost function**</font>


#### <font color="orange">Common Regression Metrics (Cost Function) </font>

##### 1. Mean Absolute Error (MAE)
The Mean Absolute Error measures the average magnitude of the errors in a set of predictions, without considering their direction. It gives an idea of how much the predicted values deviate from the actual values.

$$
MAE = \frac{1}{n} \sum_{i=1}^n |y_i - \hat{y}_i|
$$
- Robust to outliers
- It will be in the same unit
- Convergence usually take more time as optimization is a coplex task


##### 2. Mean Squared Error (MSE)
The Mean Squared Error is the average of the squared differences between the actual and predicted values. It penalizes larger errors more than smaller ones, making it sensitive to outliers.

$$
MSE = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
$$

- It is diffrentiable at all point
- It has one local and one global minima
- Converges faster
- Sensitive to outliers
-  its not in the same unit



##### 3. Root Mean Squared Error (RMSE)
The Root Mean Squared Error is the square root of the Mean Squared Error. It provides an estimate of the standard deviation of the errors, making it easier to interpret in the same units as the target variable.

$$
RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2}
$$
- Same unit
- Diffrentialble
- not robust to outliers

##### 4. Coefficient of Determination (R²)
The R² score indicates the proportion of the variance in the dependent variable that is predictable from the independent variable(s). It ranges from 0 to 1, where 1 indicates perfect predictions.

$$
R^2 = 1 - \frac{\sum_{i=1}^n (y_i - \hat{y}_i)^2}{\sum_{i=1}^n (y_i - \bar{y})^2}
$$

- ** $\sum_{i=1}^n (y_i - \hat{y}_i)^2 $ ** - Sum of Square residual
- **$\sum_{i=1}^n (y_i - \bar{y})^2 $ ** - Sum of Square total

##### 5. Adjusted R²
The Adjusted R² adjusts the R² value to account for the number of predictors in the model. It prevents overestimating the model's explanatory power when additional predictors are included.

$$
R_{adj}^2 = 1 - \left( \frac{(1 - R^2)(n - 1)}{n - k - 1} \right)
$$

- n - numbers of data points
- p - number of independent features

##### <font color="green">Why R2 is not always a good metric for model evaluation? </font>

- As the numbers of independent variable increases, the value of R2 tends to increase. 
- Using adjusted R2 it penealize the model if the independent variables are not correlated with the dependent features.


##### <font color="orange">Example: Gradient Descent </font>

Gradient Descent is a popular convergence algorithm used in machine learning. Here's how it works:
- Initialize: Start with an initial guess for the parameters, θ.
- Iterate: Use the current parameters to make predictions and calculate the cost, J(θ).
- Improve: Update the parameters using the gradient of the cost function, ∇J(θ).
- Repeat: Go back to step 2 and repeat until convergence.

#### <font color="orange">Convergence Algorithm </font>

The convergence algorithm ensures that an iterative optimization process converges to a solution by repeatedly updating parameters until a stopping criterion is met.

#### <font color="orange">Convergence Algorithm Steps </font>

1. **Initialize Parameters**:
   Start with initial values for the parameters:
    $ \theta $, such as (  $ \theta_1 $,  $ \theta_2 $, $ \ldots $, $ \theta_n $).
    

2. **Compute the Objective Function**:
   Evaluate the cost or loss function $ J(\theta) $, which measures the difference between the predicted and actual values.

   Example of a cost function (Mean Squared Error):
   $$
   J(\theta) = \frac{1}{2m} \sum_{i=1}^m (h_\theta(x^{(i)}) - y^{(i)})^2
   $$

3. **Compute the Gradient**:
   Calculate the gradient of the cost function with respect to the parameters:
   $$
   \frac{\partial J(\theta)}{\partial \theta_j}
   $$

   The gradient provides the direction and magnitude of the parameter updates.

4. **Update the Parameters**:
   Update the parameters using the chosen learning rate $ \alpha $:
   $$
   \theta_j := \theta_j - \alpha \frac{\partial J(\theta)}{\partial \theta_j}
   $$

`( weight {new_parameter} = weight{old_parameter} - {learning_rate} * {gradient} )`

5. **Check the Stopping Criterion**:
   Stop the iterations when one of the following conditions is met:
   - The change in $ J(\theta) $ between iterations is less than a predefined threshold $ \epsilon $.
     $$
     |J(\theta^{(t)}) - J(\theta^{(t-1)})| < \epsilon
     $$
   - The number of iterations exceeds the maximum allowed iterations.

6. **Output the Optimal Parameters**:
   Once convergence is achieved, return the optimal parameters $ \theta^* $.


#### <font color="orange">Learning Rate </font> 

The learning rate determines how quickly or slowly the model learns from the training data. It controls the step size of each update to the model's parameters during the optimization process.

##### <font color="orange">How Does the Learning Rate Affect the Optimization Process?</font>

A high learning rate can lead to:
- Rapid convergence, but may overshoot the optimal solution, causing oscillations.
- Faster learning, but may not converge to the global minimum.

A low learning rate can lead to:
- More stable convergence, but may require more iterations to reach the optimal solution.
- Slower learning, but may converge to the global minimum.

##### <font color="orange">Choosing the Optimal Learning Rate </font>

The choice of learning rate depends on the specific problem, dataset, and model architecture. Here are some general guidelines:
- Start with a relatively high learning rate (e.g., 0.1 or 0.01) and gradually decrease it during training.
- Use a learning rate scheduler to adapt the learning rate during training.
- Monitor the model's performance on the validation set and adjust the learning rate accordingly.

##### <font color="orange">Common Learning Rate Schedulers </font>

- **Step-wise Learning Rate Schedule**: Reduce the learning rate by a fixed factor at regular intervals.
- **Exponential Learning Rate Schedule**: Reduce the learning rate exponentially over time.
- **Cosine Annealing Learning Rate Schedule**: Reduce the learning rate using a cosine annealing schedule.

##### <font color="orange">Best Practices for Setting the Learning Rate </font>

- **Start with a high learning rate**: Begin with a relatively high learning rate to quickly converge to the optimal solution.
- **Monitor the model's performance**: Keep track of the model's performance on the validation set and adjust the learning rate accordingly.
- **Use a learning rate scheduler**: Employ a learning rate scheduler to adapt the learning rate during training.
- **Avoid very small learning rates**: Very small learning rates can lead to slow convergence or getting stuck in a local minimum.


#### <font color="light blue">Multiple Linear Regression</font>

#### <font color="orange">What is Multiple Linear Regression? </font>

Multiple Linear Regression is a statistical technique that models the relationship between one dependent variable and two or more independent variables by fitting a linear equation to the observed data.

##### <font color="orange">Mathematical Equation</font>
The Multiple Linear Regression equation is:
$$
y = β0 + β1x1 + β2x2 + ... + βnxn + ε
$$
Where:
- **y**: Dependent variable
- **x1, x2, ..., xn**: Independent variables
- **β0**: Intercept
- **β1, β2, ..., βn**: Coefficients
- **ε**: Error term

### <font color="orange">Diagram</font>
![Multiple Linear Regression](https://upload.wikimedia.org/wikipedia/commons/1/1a/Linear_regression.svg)

##### <font color="orange">Assumptions</font>
1. **Linearity**: The relationship between the dependent and independent variables is linear.
2. **Independence**: Observations are independent of each other.
3. **Homoscedasticity**: Constant variance of the errors.
4. **Normality**: The errors are normally distributed.
5. **No multicollinearity**: Independent variables are not highly correlated.

##### <font color="orange">Steps to Perform Multiple Linear Regression</font>
1. **Collect Data**: Gather data on the dependent and independent variables.
2. **Preprocess Data**: Handle missing values, encode categorical variables, and scale features if necessary.
3. **Fit the Model**: Use statistical software or libraries to fit the regression model.
4. **Evaluate the Model**: Check R2, Adjusted R2, and other metrics to evaluate model performance.
5. **Interpret Results**: Analyze the coefficients and their significance.

##### <font color="orange">Common Metrics</font>
1. **R2**: Proportion of variance in the dependent variable explained by the independent variables.
2. **Adjusted R2**: Adjusted for the number of predictors in the model.
3. **MAE**: Mean Absolute Error.
4. **MSE**: Mean Squared Error.
5. **RMSE**: Root Mean Squared Error.

#### <font color="Oragne">Overfitting and Underfitting</font>

| **Aspect**       | **Overfitting**                                                                 | **Underfitting**                                                               |
|------------------|---------------------------------------------------------------------------------|--------------------------------------------------------------------------------|
| **Definition**   | Model learns the training data too well, capturing noise and outliers.          | Model is too simple to capture the underlying patterns in the data.            |
| **Symptoms**     | High accuracy on training data, low accuracy on validation/test data.           | Low accuracy on both training and validation/test data.                        |
| **Causes**       | Model is too complex, insufficient training data, too many features.            | Model is too simple, insufficient training time, too few features.             |
| **Consequences** | Poor generalization to new data, high variance.                                 | Poor performance on both training and new data, high bias.                     |
| **Solutions**    | Simplify the model, use regularization, collect more training data, use cross-validation, prune features. | Increase model complexity, train the model longer, add more relevant features, reduce noise in the data. |
| **Bias**         | Low bias.                                                                       | High bias.                                                                     |
| **Variance**     | High variance.                                                                  | Low variance.  

**Key Points to Remember:**
- **Bias-Variance Tradeoff**: Balancing between overfitting and underfitting is crucial. High bias leads to underfitting, while high variance leads to overfitting.
- **Regularization**: Techniques like L1 and L2 regularization help in preventing overfitting by adding a penalty for larger coefficients.
- **Cross-Validation**: Using cross-validation helps in assessing the model's performance on unseen data and tuning hyperparameters effectively.
- **Feature Selection**: Selecting relevant features and removing irrelevant ones can help in reducing both overfitting and underfitting.                                                                |

#### <font color="light blue">Linear Regression with Ordinary Least Squares (OLS)</font>

#### <font color="orange">What is Linear Regression?</font>

Linear Regression is a statistical method used to model the relationship between a dependent variable (target) and one or more independent variables (predictors) by fitting a linear equation to the observed data.

#### <font color="orange">Mathematical Equation</font>

The linear regression equation is:
$$
y = β0 + β1x1 + β2x2 + ... + βnxn + ε
$$
Where:
- **y**: Dependent variable (target)
- **x1, x2, ..., xn**: Independent variables (predictors)
- **β0**: Intercept (constant term)
- **β1, β2, ..., βn**: Coefficients (slopes)
- **ε**: Error term (residuals)

#### <font color="orange">Ordinary Least Squares (OLS)</font>

OLS is a method to estimate the coefficients (β0, β1, ..., βn) by minimizing the sum of the squared differences between the observed and predicted values of the dependent variable.

#### <font color="orange">Objective Function</font>

The objective function (cost function) to minimize is:
$$
J(β) = \sum_{i=1}^n (y_i - \hat{y}_i)^2
$$
Where:
- **J(β)**: Cost function
- $ y_i $: Actual value of the dependent variable
- $ \hat{y}_i $: Predicted value of the dependent variable

#### <font color="orange">Normal Equation</font>

The OLS estimates can be obtained using the normal equation:
$$
β = (X^TX)^{-1}X^Ty
$$
Where:
- **X**: Matrix of independent variables
- **y**: Vector of dependent variable
- **β**: Vector of coefficients



#### <font color="light blue">Using OLS in Multiple Linear Regression</font>

In multiple linear regression, OLS is used to estimate the coefficients (β0, β1, ..., βn) by minimizing the sum of the squared differences between the observed and predicted values of the dependent variable.

#### <font color="orange">Steps to Perform OLS in Multiple Linear Regression</font>

1. **Prepare the Data**: Organize the data into a matrix of independent variables (X) and a vector of the dependent variable (y).

2. **Add a Column of Ones**: Add a column of ones to the matrix of independent variables (X) to account for the intercept term (β0).

3. **Compute the Coefficients**: Use the normal equation to compute the coefficients (β):
    $$
    β = (X^TX)^{-1}X^Ty
    $$

4. **Make Predictions**: Use the estimated coefficients to make predictions for the dependent variable (y).

#### <font color="orange">Example Code</font>

Here's an example of how to perform OLS in multiple linear regression using Python:
```python
import numpy as np

# Sample data
X = np.array([[1, 2104, 5, 1, 45],
              [1, 1416, 3, 2, 40],
              [1, 1534, 3, 2, 30],
              [1, 852, 2, 1, 36]])
y = np.array([460, 232, 315, 178])

# Compute the coefficients using the normal equation
beta = np.linalg.inv(X.T @ X) @ X.T @ y

# Make predictions
y_pred = X @ beta

print("Coefficients:", beta)
print("Predicted values:", y_pred)
```

#### What is Standardization?
Standardization is a preprocessing technique used to transform the features of a dataset so that they have a mean of zero and a standard deviation of one. This process ensures that each feature contributes equally to the model's performance.

#### Why is Standardization Important?
1. **Improves Model Performance**: Standardization can improve the performance of machine learning algorithms, especially those that rely on distance metrics (e.g., K-Nearest Neighbors, Support Vector Machines).
2. **Speeds Up Convergence**: For optimization algorithms like gradient descent, standardization can speed up convergence by ensuring that all features are on a similar scale.
3. **Prevents Dominance**: It prevents features with larger scales from dominating the learning process, ensuring that all features contribute equally.

#### How to Standardize Data?
Standardization is typically done using the following formula:
$$
z = \frac{x - \mu}{\sigma}
$$
Where:
- **x**: Original feature value
- **μ**: Mean of the feature
- **σ**: Standard deviation of the feature
- **z**: Standardized feature value

#### Mean and Standard Deviation Relationship
- **Mean (μ)**: The average value of the feature.
- **Standard Deviation (σ)**: A measure of the amount of variation or dispersion in the feature values.

After standardization:
- The mean of the standardized feature will be 0.
- The standard deviation of the standardized feature will be 1.

##### Why use `scaler.fit_transform` for training data and `scaler.transform` for test data?

- **`scaler.fit_transform` on training data:**
  - **Fit:** Calculates the mean and standard deviation (or other statistics) from the training data.
  - **Transform:** Applies the scaling (e.g., normalization or standardization) to the training data using the calculated statistics.
  - **Purpose:** Ensures that the model is trained on data that has been scaled consistently.

- **`scaler.transform` on test data:**
  - **Transform:** Applies the scaling to the test data using the statistics calculated from the training data.
  - **Purpose:** Ensures that the test data is scaled in the same way as the training data, providing a fair evaluation of the model.

##### What is data leakage?

- **Definition:** Data leakage occurs when information from outside the training dataset is used to create the model. This can lead to overly optimistic performance estimates during training and poor generalization to new data.
- **Causes:**
  - Including test data in the training process.
  - Using future data points in time series forecasting.
  - Leakage of target variable information into the features.
- **Consequences:**
  - Inflated model performance during training.
  - Poor performance on unseen data.
- **Prevention:**
  - Strict separation of training and test datasets.
  - Careful feature engineering to avoid including future information.
  - Cross-validation techniques to ensure robust model evaluation.

#### Assumption of Linear Regression

1. **Linearity:**
   - The relationship between the independent variables (features) and the dependent variable (target) is linear.
   - This can be checked using scatter plots or residual plots.

2. **Independence:**
   - Observations are independent of each other.
   - This is particularly important in time series data where autocorrelation can be an issue.

3. **Homoscedasticity:**
   - The residuals (errors) have constant variance at every level of the independent variables.
   - This can be checked using residual plots (residuals vs. fitted values).

4. **No Multicollinearity:**
   - Independent variables are not highly correlated with each other.
   - This can be checked using Variance Inflation Factor (VIF) or correlation matrices.

5. **Normality of Residuals:**
   - The residuals are normally distributed.
   - This can be checked using Q-Q plots or histograms of the residuals.

6. **No Autocorrelation:**
   - Residuals are not correlated with each other.
   - This can be checked using the Durbin-Watson test.

7. **No Endogeneity:**
   - There is no correlation between the independent variables and the error term.
   - This can be checked using statistical tests or domain knowledge.

It is a form of Linaer Regression technique where the relationship between the independent variable (features) and the dependent variable (target) is modeled as an **nth degree polynomial**.

## How does a Polynomial Regression work?

When the relationship is non-linear, a polynomial regression model introduces higher-degree polynomial terms. By including higher-degree terms (quadratic, cubic, etc.), the model can capture the non-linear patterns in the data.

The polynomial regression equation for degree n is given by:

Polynomial regression is a form of regression analysis in which the relationship between the independent variable \( x \) and the dependent variable \( y \) is modeled as an \( n \)-th degree polynomial.

The general form of a polynomial regression model is:

$$ \ y = \beta_0 + \beta_1 x + \beta_2 x^2 + \beta_3 x^3 + \cdots + \beta_n x^n + \epsilon \ $$

Where:
- \( y \) is the dependent variable.
- \( x \) is the independent variable.
- \($ \beta_0 $, $ \beta_1 $, $ \beta_2 $ , $ \ldots $, $ \beta_n  $\) are the coefficients of the polynomial.
- \( n \) is the degree of the polynomial.
- \( $\epsilon $\) is the error term.

##### Key Points

- **Degree of Polynomial:** The degree \( n \) determines the complexity of the model. Higher degrees can fit the training data better but may lead to overfitting.
- **Feature Scaling:** Polynomial features can have large ranges, so feature scaling may be necessary.
- **Regularization:** Techniques like Ridge or Lasso regression can be used to prevent overfitting in polynomial regression.

````python
# Creating a function for polynomial regression
def polynomial_regression(degree):
    from sklearn.pipeline import Pipeline
    from sklearn.preprocessing import PolynomialFeatures
    from sklearn.linear_model import LinearRegression
    
    X_new = np.linspace(X.min(), X.max(), 200).reshape(200, 1)
    poly_features = PolynomialFeatures(degree=degree, include_bias=False)
    lin_reg = LinearRegression()
    poly_regression = Pipeline([
        ('poly_features', poly_features),
        ('linear_reg', lin_reg)
    ])
    poly_regression.fit(X_train, y_train)
    y_pred_new = poly_regression.predict(X_new)
    
    return X_new, y_pred_new
````
